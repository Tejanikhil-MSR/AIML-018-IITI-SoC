```python
import os
import json

def convert_json_to_ndjson(json_file, ndjson_file):
    with open(json_file, 'r', encoding='utf-8') as f:
        data = json.load(f)

    with open(ndjson_file, 'w', encoding='utf-8') as f:
        if isinstance(data, list):
            for item in data:
                f.write(json.dumps(item) + '\n')
        elif isinstance(data, dict):
            f.write(json.dumps(data) + '\n')
        else:
            raise ValueError("Unsupported JSON root type")

def convert_all_json_to_ndjson(root_dir):
    for dirpath, _, filenames in os.walk(root_dir):
        for filename in filenames:
            if filename.endswith(".json"):
                json_path = os.path.join(dirpath, filename)
                ndjson_path = json_path.replace(".json", ".ndjson")
                try:
                    convert_json_to_ndjson(json_path, ndjson_path)
                    os.remove(json_path)  # Remove original JSON file
                    print(f"[✔] Converted and deleted: {json_path} → {ndjson_path}")
                except Exception as e:
                    print(f"[✘] Failed to convert {json_path}: {e}")



# Example: call with your directory path
convert_all_json_to_ndjson("/home/saranshvashistha/workspace/AIML-018-IITI-SoC/data")
```

```python
from pathlib import Path

def clean_text_file(input_path: str, output_path: str = None):
    input_path = Path(input_path)
    if output_path is None:
        output_path = input_path.with_name(f"{input_path.stem}_cleaned.txt")

    with input_path.open("r", encoding="utf-8", errors="ignore") as f:
        content = f.read()

    # Optional: remove non-printable characters (e.g., control codes)
    cleaned_content = ''.join(c for c in content if c.isprintable() or c in '\n\r\t')

    with open(output_path, "w", encoding="utf-8") as f:
        f.write(cleaned_content)

    print(f"[✔] Cleaned file saved to: {output_path}")

for dirpath, _, filenames in os.walk("/home/saranshvashistha/workspace/AIML-018-IITI-SoC/text_op"):
        for filename in filenames:
            if  filename.endswith(".txt") :
                filepath = os.path.join(dirpath, filename)
                clean_text_file(filepath)
```

```python
import os
import shutil

def flatten_and_copy(src_root, dst_folder):
    os.makedirs(dst_folder, exist_ok=True)

    for dirpath, _, filenames in os.walk(src_root):
        for filename in filenames:
            src_path = os.path.join(dirpath, filename)
            dst_path = os.path.join(dst_folder, filename)

            if os.path.exists(dst_path):
                print(f"Skipping (already exists): {filename}")
                continue

            shutil.copy2(src_path, dst_path)
            print(f"Copied: {src_path} → {dst_path}")

# Example usage:
flatten_and_copy("/home/saranshvashistha/workspace/AIML-018-IITI-SoC/text_op", "/home/saranshvashistha/workspace/AIML-018-IITI-SoC/data_clean_total")
```

```python
import os

def count_files_in_folder(folder_path):
    return sum(1 for entry in os.scandir(folder_path) if entry.is_file())

# Example usage:
count = count_files_in_folder("/home/saranshvashistha/workspace/AIML-018-IITI-SoC/data_clean_total")
print(f"Total files in folder: {count}")
```

```python
import os
import shutil
import random

def copy_random_files(src_folder, dst_folder, n=50):
    os.makedirs(dst_folder, exist_ok=True)

    # Get list of all files in the source folder (non-recursive)
    files = [f for f in os.listdir(src_folder) if os.path.isfile(os.path.join(src_folder, f))]

    # Sample n files (or all if fewer than n)
    selected_files = random.sample(files, min(n, len(files)))

    for file_name in selected_files:
        src_path = os.path.join(src_folder, file_name)
        dst_path = os.path.join(dst_folder, file_name)
        shutil.copy2(src_path, dst_path)
        print(f"Copied: {file_name}")

# Example usage:
copy_random_files("/home/saranshvashistha/workspace/AIML-018-IITI-SoC/data_clean_total", "/home/saranshvashistha/workspace/AIML-018-IITI-SoC/data_clean_50", n=200)
```

```python
import os

def print_file_sizes(folder_path):
    for entry in os.scandir(folder_path):
        if entry.is_file():
            size_bytes = os.path.getsize(entry.path)
            size_kb = size_bytes / 1024
            print(f"{entry.name} — {size_kb:.2f} KB")

# Example usage:
print_file_sizes("/home/saranshvashistha/workspace/AIML-018-IITI-SoC/data_clean_total")
```
